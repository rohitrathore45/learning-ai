```python
import numpy as np
import pandas as pd
from pyparsing import col
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import seaborn as sns
```

```python
df = pd.read_csv("Titanic-Dataset.csv")[['Age', 'Pclass', 'SibSp', 'Parch', 'Survived']]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
X = df.iloc[:, 0:4]
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(), X, y, cv=20, scoring='accuracy'))
```

Applying Feature Construction

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1
```

```python
X.head()
```

```python
def myfunc(num):
    if num == 1:
        return 0
    elif num > 1 and num <= 4:
        return 1
    else:
        return 2
```

```python
myfunc(4)
```

```python
X['Family_type'] = X['Family_size'].apply(myfunc)
```

```python
X.head()
```

```python
X.drop(columns = ['Family_size', 'Parch', 'SibSp'], inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(), X, y, cv=20, scoring='accuracy'))
```
