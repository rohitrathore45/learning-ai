```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.utils.extmath import density
```

```python
df = pd.read_csv("data_science_job.csv")
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
df.shape
```

```python
cols = [var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean() > 0]
cols
```

```python
df[cols].sample(5)
```

```python
len(df[cols].dropna()) / len(df)
```

```python
new_df = df[cols].dropna()
df.shape, new_df.shape
```

```python
import matplotlib.pyplot as plt
new_df.hist(bins = 50, density = True, figsize = (12,12))
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['training_hours'].hist(bins = 50, ax = ax, density = True, color = 'r')

new_df['training_hours'].hist(bins = 50, ax = ax, density = True, color = 'b', alpha = 0.8)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['city_development_index'].hist(bins = 50, ax = ax, density = True, color = 'r')

new_df['city_development_index'].hist(bins = 50, ax = ax, density = True, color = 'b', alpha = 0.8)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['experience'].hist(bins = 50, ax = ax, density = True, color = 'r')

new_df['experience'].hist(bins = 50, ax = ax, density = True, color = 'b', alpha = 0.8)
```
