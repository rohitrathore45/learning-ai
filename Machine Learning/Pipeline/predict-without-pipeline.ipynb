```python
import pickle
import numpy as np
import pandas as pd
```

```python
ohe_sex = pickle.load(open('model/ohe_sex.pkl', 'rb'))
ohe_embarked = pickle.load(open('model/ohe_embarked.pkl', 'rb'))
clf = pickle.load(open('model/clf.pkl', 'rb'))
```

```python
# assume user input
# Pclass/gender/age/SibSp/Parch/Fare/Embarked

test_input = np.array([2, 'male', 31.0, 0, 0, 10.5, 'S'], dtype= object).reshape(1, 7)
```

```python
test_input
```

```python
test_input_sex = ohe_sex.transform(pd.DataFrame(test_input[:, 1], columns=['Sex']))
```

```python
test_input_sex
```

```python
test_input_embarked = ohe_embarked.transform(test_input[:, -1].reshape(1, 1))
```

```python
test_input_embarked
```

```python
test_input_age = test_input[:, 2].reshape(1, 1)
test_input_age
```

```python
test_input_transformed = np.concatenate((test_input[:, [0, 3, 4, 5]], test_input_age, test_input_embarked, test_input_sex), axis=1)
```

```python
test_input_transformed.shape
```

```python
clf.predict(test_input_transformed)
```
