```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.iloc[:, 3:5].values
```

```python
df
```

```python
import scipy.cluster.hierarchy as hc

plt.figure(figsize=(10,7))
plt.title('Hierarchical Clustering Dendrogram')
dend = hc.dendrogram(hc.linkage(df, method='ward'))
plt.grid(False)
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, linkage='ward')
labels_ = cluster.fit_predict(df)
```

```python
labels_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(df[:, 0], df[:, 1], c = cluster.labels_, cmap = 'rainbow')
plt.grid(False)
```
