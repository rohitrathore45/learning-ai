```python
import matplotlib.pyplot as plt

x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
y = [10, 7, 12, 15, 20, 17, 14]
plt.plot(x, y)
plt.title("Bakery sales this week!")

plt.xlabel("Days of week")
plt.ylabel("Sales per day")
plt.show()
```
