```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data = pd.read_csv("vgsales.csv")

#Univariate Analysis -> Single Variable
data.head()

#Categorical
data['Genre'].unique()

cat_counts = data['Genre'].value_counts()

#Bar chart
X_bar = cat_counts.index
Y_bar = cat_counts.values

plt.figure(figsize=(12,8))
plt.bar(X_bar, Y_bar)
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90, fontsize = 10)
plt.bar(X_bar, Y_bar, width=0.5, color = 'teal')
plt.xlabel('Genre', fontsize = 10)
plt.ylabel('Count', fontsize = 10)
plt.title('Top Genre Analysis', fontsize = 10)
plt.show()
```

```python
sns.countplot(x='Genre', data=data, order = data['Genre'].value_counts().index, color = 'orange')
```

```python
sales_data = data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
sales_data
```

```python
region_sales = sales_data.T.sum(axis='columns')
region_sales
```

```python
plt.pie(region_sales, labels=region_sales.index, startangle=90, autopct='%1.1f%%', explode=(0.1, 0, 0, 0))
plt.title('Total Sales by Region', fontsize = 10)
plt.show()
```

```python
top_5_pub = data['Publisher'].value_counts().index[:5]
```

```python
top_5_genre = data['Genre'].value_counts().index[:5]
```

```python
top_5_plat = data['Platform'].value_counts().index[:5]
```

```python
top_5_plat
```

```python
top_5_data = data.loc[(data['Publisher'].isin(top_5_pub)&(data['Platform'].isin(top_5_plat))&(data['Genre'].isin(top_5_genre)))]
```

```python
top_5_data
```

```python
sns.countplot(data = top_5_data, x = 'Publisher', hue = 'Platform')
plt.xticks(rotation=90, fontsize = 10)
plt.legend(bbox_to_anchor=(-0.2, 1))
plt.show()
```

```python
df_stacked_plt= pd.crosstab(index = top_5_data['Publisher'], columns = top_5_data['Platform'])
df_stacked_plt
```

```python
df_stacked_plt.plot(kind='bar', stacked=True)
plt.legend(bbox_to_anchor=(-0.2, 1))
plt.grid(False)
plt.show()
```

```python
sns.boxplot(x = 'Publisher', y = 'Global_Sales', data = top_5_data)
plt.xticks(rotation=90, fontsize = 10)
```
